```python
#Reading the dataset 
import pandas as pd
df = pd.read_csv('../ecommerce_customer.csv')
```

```python
# Display the first few rows of the dataframe
df.head()
```

```python
#checking the all columns datatypes
df.info()
```

```python
# Check for missing values in the dataset
df.isnull().sum()
```

```python
# separating numerical and categorical columns

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
df.head(2)
```

```python
# Filling with median values for rate_columns
rate_columns = ["Cart_Abandonment_Rate",
    "Returns_Rate",
    "Email_Open_Rate",
    "Discount_Usage_Rate"]

for col in rate_columns:
    df[col] = df[col].fillna(df[col].median())
```

```python
num_col = [
    "Session_Duration_Avg","Days_Since_Last_Purchase",
    "Customer_Service_Calls","Social_Media_Engagement_Score",
    "Mobile_App_Usage","Payment_Method_Diversity","Credit_Balance"
]
for col in num_col:
    df[col] = df[col].fillna(df[col].median())
```

```python
num_cols = [
    "Wishlist_Items","Product_Reviews_Written","Pages_Per_Session"
]
df[num_cols] = df[num_cols].fillna(0)
    
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())

# Verify that there are no missing values left
df.isnull().sum()
```

```python
# Check for duplicate rows
df = df.drop_duplicates()
# Check how many duplicate rows were removed
df.duplicated().sum()
```

```python
#Removing unnecessary columns
columns_to_remove = [
    "City",
    "Signup_Quarter",
    "Lifetime_Value"
]

df.drop(columns=columns_to_remove, inplace=True)
```
